```python
from bs4 import BeautifulSoup
import requests
for i in range(1,3):
    url = BeautifulSoup(f'http://www.nepalstock.com/main/floorsheet/index/{i}/','html.parser')

    response = requests.get(url) # requesting for the website
    data = response.text          # converting into string formate
    soup = BeautifulSoup(data,'lxml') # changing into lxml 


    tags = soup.table
    tag = tags.find_all('tr', align = False)

    list1 = []
    for i in range(len(tag)):
        x = tag[i].get_text()
        list1.append(x)

    list2 = []
    for i in range(len(list1)):
        x = list1[i].split("\n")
        list2.append(x[1:-1])
  #list2
    list2 =  list2[0:-3]

    import csv
    xlen = len(list2)
    with open("StockPrice1.csv","a") as csvFile:
    for i in range(xlen):
        writer = csv.writer(csvFile)
        writer.writerow(list2[i])
```
